```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_csv('lis_data.csv')

df.head()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.lineplot(x='year',y='poorAll6',hue='country',markers=True,data=df,ax=ax)
```

```python
data = df.set_index(['country','year'])

fig, ax = plt.subplots(figsize=(8,5))

data.loc['United Kingdom','poorAll6'].plot(marker='o',label='United Kingdom',ax=ax)
data.loc['United States','poorAll6'].plot(marker='o',label='United States',ax=ax)

plt.legend(loc='upper left',bbox_to_anchor=(1,1),title='Country')
plt.xlabel('Year')
plt.text(1976,15.25,'Source: Luxembourg Income Study')
plt.title('Relative poverty in the US and the UK')
plt.ylabel('Share below 60% of median income')
plt.savefig('rel-pov-all.png',bbox_inches='tight',dpi=250)
```

```python
data = df.set_index(['country','year'])

fig, ax = plt.subplots(figsize=(8,5))

data.loc['United Kingdom','poorK6'].plot(marker='o',label='United Kingdom',ax=ax)
data.loc['United States','poorK6'].plot(marker='o',label='United States',ax=ax)

plt.legend(loc='upper left',bbox_to_anchor=(1,1),title='Country')
plt.xlabel('Year')
plt.text(1980,12.25,'Source: Luxembourg Income Study')
plt.title('Relative child poverty in the US and the UK')
plt.ylabel('Share below 60% of median income')
plt.savefig('rel-pov-children.png',bbox_inches='tight',dpi=250)
```
