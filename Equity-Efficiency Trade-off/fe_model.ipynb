```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from linearmodels import PanelOLS

gini = pd.read_csv('swiid8_3_summary.csv')[['country','year','gini_disp']]
prod = pd.read_stata('pwt91.dta')[['country','year','rgdpe','emp','hc','csh_i','csh_x']]
prod['y'] = np.log(prod['rgdpe']/prod['emp'])
prod = prod[['country','year','y','hc','csh_i','csh_x']].dropna(subset=['y'])

df = pd.merge(gini,prod)

df['g'] = df.groupby('country')['y'].diff().shift(-1)
df.dropna(subset=['g'],inplace=True)
df['gini_disp']*=1/100
df = df.set_index(['country','year'])
df.head()
```

```python
res = PanelOLS.from_formula('g ~ 1 + y + hc + csh_i + csh_x + gini_disp + TimeEffects + EntityEffects',data=df)
res.fit()
```
