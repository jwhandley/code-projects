```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_stata('BES2019_W16_v0.2.dta',columns=['lr_scale','al_scale','profile_past_vote_2017','profile_eurefvote','wt_new_W13_result','al1','al2','al3','al4','al5','lr1','lr2','lr3','lr4','lr5'])

df['left_right'] = df['lr_scale'].cat.codes.replace(-1,np.nan)/2
df['authoritarian_liberal'] = df['al_scale'].cat.codes.replace(-1,np.nan)/2

#df = pd.read_stata('BES2019_W16_v0.2.dta',convert_categoricals=False)

df.head()
```

```python
for i in ['al{}'.format(i) for i in np.arange(1,6)]:
    df[i+'_code'] = (df[i].cat.codes+1).replace(6,np.nan)
    
for i in ['lr{}'.format(i) for i in np.arange(1,6)]:
    df[i+'_code'] = (df[i].cat.codes+1).replace(6,np.nan)
    
pd.pivot_table(df,index='lr2',values='lr2_code')
```

```python
import statsmodels.formula.api as smf

res = smf.ols('left_right ~ lr1_code + lr2_code + lr3_code + lr4_code + lr5_code',data=df).fit()
res.summary()
```

```python
res1 = smf.ols('authoritarian_liberal ~ al1_code + al2_code + al3_code + al4_code + al5_code',data=df).fit()
res1.summary()
```
