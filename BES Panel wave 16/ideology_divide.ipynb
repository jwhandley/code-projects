```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_stata('BES2019_W16_v0.2.dta',columns=['lr_scale','al_scale','profile_past_vote_2017','profile_eurefvote','wt_new_W13_result'])

df['left_right'] = df['lr_scale'].cat.codes.replace(-1,np.nan)/2
df['authoritarian_liberal'] = df['al_scale'].cat.codes.replace(-1,np.nan)/2

#df = pd.read_stata('BES2019_W16_v0.2.dta',convert_categoricals=False)

df.head()
```

```python
pd.pivot_table(df,index='profile_past_vote_2017',columns='profile_eurefvote',values=['left_right','authoritarian_liberal']).loc[['Conservative','Labour']]
```

```python
df['wt_lr'] = df['wt_new_W13_result']*df['left_right']
df['wt_al'] = df['wt_new_W13_result']*df['authoritarian_liberal']

wgt = pd.pivot_table(df,index='profile_past_vote_2017',columns='profile_eurefvote',values=['wt_lr','wt_al'],aggfunc='sum').loc[['Conservative','Labour']]
pop = pd.pivot_table(df,index='profile_past_vote_2017',columns='profile_eurefvote',values=['wt_new_W13_result'],aggfunc='sum').loc[['Conservative','Labour']]

table = wgt/pop['wt_new_W13_result']
table
```

```python
print('Authoritarian-Liberal')
print(table['wt_al']['Stay/remain in the EU']-table['wt_al']['Leave the EU'])

print('Left-Right')
print(table['wt_lr']['Stay/remain in the EU']-table['wt_lr']['Leave the EU'])
```
