```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_stata('mpd2018.dta')
df['year'] = df['year'].astype(int)
df.set_index(['countrycode','year'],inplace=True)

reg = pd.read_stata('mpd2018_region_data.dta')
reg['year'] = reg['year'].astype(int)
reg.set_index(['region','year'],inplace=True)

reg.head()
```

```python
reg.loc['we','cgdppc'].plot(label='Western Europe')
df.loc['USA','cgdppc'].plot(label='United States')
df.loc['JPN','cgdppc'].plot(label='Japan')
plt.legend()
plt.axis([1870,2020,0,60000])
```
