```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_stata('pwt91.dta')
df['y'] = df['rgdpo']/(df['emp']*df['avh'])
df['yna'] = df['rgdpna']/(df['emp']*df['avh'])
df['yc'] = df['cgdpo']/(df['emp']*df['avh'])

df['kc'] = df['ck']/(df['emp']*df['avh'])
df['rk'] = df['rkna']/(df['emp']*df['avh'])
data = pd.pivot_table(df,index=['country','year'],values=['y','yna','yc','kc','rk'])

data.head(6)
```

```python
(data.loc['United Kingdom','yc']/data.loc['United States','yc']).plot(label='United Kingdom')
(data.loc['France','yc']/data.loc['United States','yc']).plot(label='France')
(data.loc['Germany','yc']/data.loc['United States','yc']).plot(label='Germany')
(data.loc['Canada','yc']/data.loc['United States','yc']).plot(label='Canada')
plt.legend()
```

```python
anglo = ['United Kingdom','Australia','New Zealand','Canada','United States','Ireland']
eu6 = ['Belgium','France','Italy','Germany','Netherlands','Luxembourg']

np.log(data.loc[anglo,'y'].unstack(0).median(axis=1)).plot(label='Anglo-Saxon Economies')
np.log(data.loc[eu6,'y'].unstack(0).median(axis=1)).plot(label='EU6')
np.log(data.loc['United Kingdom','y']).plot(label='United Kingdom')
plt.legend()
```

```python
(data.loc['United Kingdom','kc']/data.loc['United States','kc']).plot(label='United Kingdom')
(data.loc['France','kc']/data.loc['United States','kc']).plot(label='France')
(data.loc['Germany','kc']/data.loc['United States','kc']).plot(label='Germany')
(data.loc['Canada','kc']/data.loc['United States','kc']).plot(label='Canada')
plt.legend()
```
