```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_stata('pwt91.dta')
df['rgdpnapc'] = df['rgdpna']/df['pop']
df['rgdppc'] = df['rgdpo']/df['pop']
df['cgdppc'] = df['cgdpo']/df['pop']
df.set_index(['countrycode','year'],inplace=True)

df.head()
```

```python
df.loc['JPN',['rgdpnapc','cgdppc','rgdppc']].plot()

df.loc['GBR',['rgdpnapc','cgdppc','rgdppc']].plot()
```

```python
df.loc['GBR','rgdpnapc'].plot(label='United Kingdom')
df.loc['DEU','rgdpnapc'].plot(label='Germany')
df.loc['FRA','rgdpnapc'].plot(label='France')
df.loc['JPN','rgdpnapc'].plot(label='Japan')
df.loc['USA','rgdpnapc'].plot(label='United States')
plt.legend()

plt.figure(2)
df.loc['GBR','rgdppc'].plot(label='United Kingdom')
df.loc['DEU','rgdppc'].plot(label='Germany')
df.loc['FRA','rgdppc'].plot(label='France')
df.loc['JPN','rgdppc'].plot(label='Japan')
df.loc['USA','rgdppc'].plot(label='United States')
plt.legend()
```

```python
np.log(df.loc['JPN','rgdpnapc']).diff().plot()
np.log(df.loc['JPN','rgdpnapc']).diff().rolling(5,center=True).mean().plot()
plt.axvspan(2001,2006,color='grey',alpha=0.5)
plt.text(1985,0.06,'Koizumi boom')
```

```python
np.log(df.loc['GBR','rgdpnapc']).diff().plot()
np.log(df.loc['GBR','rgdpnapc']).diff().rolling(5,center=True).mean().plot()
```

```python
df.loc['GBR','rgdppc'].plot(label='United Kingdom')
df.loc['JPN','rgdppc'].plot(label='Japan')
df.loc['KOR','rgdppc'].plot(label='Korea')
df.loc['TWN','rgdppc'].plot(label='Taiwan')
plt.legend()
plt.title('Real GDP per capita at purchasing power parity')
```

```python
np.log(df.loc['GBR',['rgdpo','rgdpe','rgdpna']].div(df.loc['GBR','emp']*df.loc['GBR','avh'],axis=0)).plot()
plt.legend(['Output','Expenditure','National accounts'])
plt.ylabel('Log GDP per hour worked')
plt.title('UK productivity by accounting method')
plt.text(1984.5,2.15,'Source: Penn World Table 9.1')
```
