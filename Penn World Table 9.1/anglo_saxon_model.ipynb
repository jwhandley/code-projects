```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_stata('pwt91.dta')#.set_index(['country','year'])

df['prod'] = df['rgdpe']/(df['avh']*df['emp'])

df.head()
```

```python
sns.lineplot(x='year',y='prod',hue='country',data=df.loc[df['countrycode'].isin(['GBR','USA','CAN','AUS','NZL','FRA'])])
plt.legend(loc='upper left',bbox_to_anchor=(1,1),title='Country')
plt.ylabel('GDP per hour worked')
plt.title('Productivity in Anglo-Saxon Economies vs. France')
plt.savefig('anglo_saxon_productivity.png',dpi=200,bbox_inches='tight')
```
