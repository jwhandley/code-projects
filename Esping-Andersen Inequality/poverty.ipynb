```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_csv('lis_poverty.csv')
df = df[df['year']>1980]

liberal = ['AU','CA','US','UK','IE','NZ','CH']
conservative = ['DE','AT','BE','FR','NL','IT','JP']
socdem = ['SE','DK','NO','FI']
df['regime'] = df['country'].replace(liberal+conservative+socdem,7*['liberal']+7*['conservative']+4*['social democratic'])
df = df[df['regime'].isin(['liberal','conservative','social democratic'])]
df['regime'] = df['regime'].astype('category')
df['regime'] = df['regime'].cat.reorder_categories(['liberal','conservative','social democratic'])

df.head()
```

```python
import statsmodels.formula.api as smf
from linearmodels import BetweenOLS

res = BetweenOLS.from_formula('poorAll6 ~ 1 + regime',data=df.set_index(['country','year'])).fit()
res.summary
```
