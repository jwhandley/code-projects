```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_csv('Reed-Smith-JHRED-CANDIDATES.csv')

df.head()
```

```python
pd.pivot_table(df.loc[df['result']>0],index='legis',values=['union','agri','educ','news','pubcorp']).plot()
plt.yticks(np.arange(0,0.2,0.025),['{}%'.format(i) for i in np.arange(0,20,2.5)])
plt.legend(['Agriculture','Education','News','Public corporation','Trade union'],title='Background',loc='upper left',bbox_to_anchor=(1,1))
plt.title('Share of Japanese MPs with different backgrounds')
```
