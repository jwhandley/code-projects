```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_csv('MPDataset_MPDS2019b.csv')
df['date'] = pd.to_datetime(df['edate'])
df['edate'] = pd.to_datetime(df['edate'])
df['year'] = pd.DatetimeIndex(df['date']).year
df.set_index('year',inplace=True)

df.head()
```

```python
df.loc[(df['countryname']=='Canada')&(df['partyabbrev']=='LP'),'rile'].plot(marker='o',color='C3',label='Liberal Party of Canada')
df.loc[(df['countryname']=='Canada')&(df['partyabbrev']=='NDP'),'rile'].plot(marker='o',color='C1',label='New Democratic Party')
df.loc[(df['countryname']=='United States')&(df['partyabbrev']=='Democrats'),'rile'].plot(marker='o',label='Democratic Party')
pd.pivot_table(df.loc[df['parfam']==30],index='year',values='rile',aggfunc='mean')['rile'].rolling(10).mean().plot(label='Average Social Democratic Party',color='k')
plt.legend(loc='upper left',bbox_to_anchor=(1,1),title='Party')
```
