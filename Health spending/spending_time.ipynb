```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
from linearmodels import BetweenOLS
from linearmodels import PanelOLS

exp = pd.read_csv('health-expenditure-and-financing-per-capita.csv').set_index(['iso','year'])
gdp = pd.read_stata('pwt91.dta').set_index(['countrycode','year'])

data = pd.DataFrame(index=exp.index,columns=['health_exp','gdppc','cons'])
data['health_exp'] = np.log(exp['health_exp'])
data['gdppc'] = np.log(gdp['rgdpe']/gdp['pop'])
data['cons'] = np.log(gdp['ccon']/gdp['pop'])
#data = data.reset_index()
data.to_csv('health_exp_gdp_oecd.csv')

df = pd.read_csv('health_exp_gdp_oecd.csv').set_index(['iso','year'])
df.head()
```
