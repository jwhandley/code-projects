```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_csv('oecd_2017_benchmark.csv')

df.head()
```

```python
table = pd.pivot_table(df,index=['unit','country'],columns='measure',values='values')

table
```

```python
sns.lmplot(x='gdp',y='health',data=np.log(table.loc['ppp']))
plt.scatter(np.log(51950),np.log(7413))
plt.xlabel('Log GDP per capita at PPP')
plt.ylabel('Log Health spending per capita at PPP')
plt.text(9.95,6.65,'Source: OECD 2017 PPP benchmark')
plt.savefig('health_spending_gdp.png',dpi=200)
```

```python
sns.lmplot(x='aic',y='health',data=np.log(table.loc['ppp']))
plt.scatter(np.log(39439),np.log(7413))
plt.xlabel('Log AIC per capita at PPP')
plt.ylabel('Log Health spending per capita at PPP')
plt.text(9.575,6.65,'Source: OECD 2017 PPP benchmark')
plt.savefig('health_spending_aic.png',dpi=200)
```
