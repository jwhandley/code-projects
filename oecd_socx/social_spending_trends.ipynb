```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()

df = pd.read_csv('social-expenditure-as-percentage-of-gdp.csv')
df.set_index(['iso','year'],inplace=True)
df.dropna(how='any',inplace=True)
cats = ['family','health','housing','incapacity','old_age','other','survivor','unemp','alp']
cats1 = ['family','housing','incapacity','other','survivor','unemp','alp']
df['total'] = df[cats].sum(axis=1)

df.head()
```

```python
df.loc['USA',cats1].plot.bar(stacked=True)
plt.legend(['Family','Housing','Incapacity','Other','Survivor','Unemployment','Active labour market programmes'],title='Category',loc='upper left',bbox_to_anchor=(1,1))
plt.title('Public social spending in the US excluding old age and health')
plt.ylabel('percent of GDP')
plt.savefig('us_social_spending.png',dpi=200,bbox_inches='tight')
```
